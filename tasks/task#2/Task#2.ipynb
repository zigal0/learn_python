```python
import math
class ComplexNumber:
    def __init__(self, value1, value2, form):
        if form == 'a':
            self.real = value1
            self.img = value2
        elif form == 'p':
            real = value1 * Math.cos(value2)
            img = value1 * Math.sin(value2)
    def conjugate(self):
        return ComplexNumber(self.real, -self.img, 'a')
    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.img + other.img, 'a') 
    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.img - other.img, 'a')
    def __mul__(self, other):
        return ComplexNumber(self.real * other.real - self.img * other.img, 
                             self.img * other.real + other.img * self.real, 'a')
    def __floordiv__(self, other):
        tmp = self * other.conjugate()
        return ComplexNumber(tmp.real / abs(other), tmp.img / abs(other), 'a')
    def get_algebraic_form(self):
        return (self.real, self.img)
    def get_polar_form(self):
        return (abs(self), self.arg())
    def __abs__(self):
        return math.sqrt(self.real ** 2 + self.img ** 2)
    def arg(self):
        return math.atan(self.img / self.real)
    def __eq__(self, other):
        return self.real == other.real and self.img == other.img
    def __ne__(self, other):
        return not self.__eq__(other)
    def __str__(self):
        return '(%g, %g)' % (self.real, self.img)
    def __repr__(self):
        return 'Complex' + str(self)
num1 = ComplexNumber(1, 2, 'a')
num2 = ComplexNumber(5, 7, 'a')
print((num1 + num2).get_algebraic_form())
print((num1 - num2).get_algebraic_form())
print((num1 * num2).get_algebraic_form())
print(abs(num1))
print((num1 // num2).get_algebraic_form())
# print(num1.get_algebraic_form())
# print(num1.get_polar_form())
```

```python
class Vector:
    def __init__(self, vector_values_list):
        self._coord = vector_values_list
    def __add__ (self, other):
        if len(self) != len(other):
            raise ValueError
        res = []
        for i in range(len(self)):
            res.append(self._coord[i] + other._coord[i])
        return Vector(res)
    def __mul__(self, num):
        return Vector(list(map(lambda x: x * num, self._coord)))
    def __floordiv__(self, num): 
        return self * (1 / num)
    def __sub__(self, other):
        return self + (other * -1)
    def __len__(self):
        return len(self._coord)
    def __str__(self):
        return str(self._coord)
    def __repr__(self):
        return 'Vector: ' + str(self._coord)
    def __abs__(self):
        res = 0
        for e in self._coord:
            res += e ** 2
        return math.sqrt(res)
    def __neg__(self):
        return self * -1
    def __eq__(self, other):
        return self._coord == other._coord 
    def __ne__(self, other):
        return not self.__eq__(other)
    def scalar_product(self, other):
        if len(self) != len(other):
            raise ValueError
        res = 0
        for i in range(len(self)):
            res += self._coord[i] * other._coord[i]
        return res
    def cos(self, other):
        return self.scalar_product(other) / (abs(self) * abs(other))
v1 = Vector([1, 2, 3, 4, 5])
v2 = Vector([6, 7, 8, 9, 0])
print(v1.scalar_product(v2))
print(v1.cos(v2))
print(v1)
```

```python
def calls_counter(func):
    def __init__(self):
        count = 0
    def __count(*args, **kw):
        __count.count += 1
        print('This function was called {} times'.format(__count.count))
        result = func(*args, **kw)
        return result
    __count.count = 0
    return __count
@calls_counter
def func():
    print('Hello, world!')
func()
func()
```

```python
def copy_class_attrs(cls):
    def _decorate(self):
        for f in cls.__dict__:
            if not f.startswith("_"):
                if hasattr(self, f):
                    print(f + ' 1st')
                    setattr(self, '_new' + f, getattr(cls, f))
                else:
                    print(f + ' 2nd')
                    setattr(self, f, getattr(cls, f))
        return self
    return _decorate
class MyClass:
    first = 10
    second = 1
    third = 20
    def kek_fucntion(self):
        print('Hello')
@copy_class_attrs(MyClass)
class MyClass1:
    first = 20
    second = 2  
print(dir(MyClass()))
print(dir(MyClass1()))
MyClass1().kek_fucntion()
```

```python
from copy import deepcopy
import pickle

def check_float(potential_float):
        try:
            float(potential_float)
            return True
        except ValueError:
            return False

class VersionMixin:
    def set_version(self, version):
        self._version = version
    
    @property  # One more example of decorator
    def version(self):
        return self._version if hasattr(self, '_version') else 'Unk. version'
    
class CopyMixin:
    def make_copy(self):
        return deepcopy(self)
    
class SaveTextMixin:
    def write_to_file_text(self, file_to_write):
        with open(file_to_write, 'w') as fout:
            fout.write(str(self))
            
class LoadTextMixin:
    @staticmethod
    def read_from_file_text(file_from_read):
        with open(file_from_read, 'r') as fin:
            res = []
            for line in fin.readlines():
                line = line.strip()
                cur = line.split(' ')
                cur_to_list = []
                for i in cur:
                    if i.isdigit():
                        cur_to_list.append(int(i))
                    elif check_float(i):
                        cur_to_list.append(float(i))
                res.append(cur_to_list)
            return Matrix(res)
            
class SaveBinMixin:
    def write_to_file_bin(self, file_to_write):
        with open(file_to_write, 'wb') as fout:
            pickle.dump(self, fout)
            
class LoadBinMixin:
    @staticmethod
    def read_from_file_bin(file_from_read):
        with open(file_from_read, 'rb') as fin:
            return pickle.load(fin)

class Matrix(VersionMixin, CopyMixin, SaveTextMixin, SaveBinMixin, LoadBinMixin, LoadTextMixin):
    def __init__(self, *args):
        if len(args) == 1:
            if (isinstance(args[0], list)):
                if len(args[0]) == 0:
                    self._matrix = [[]]
                    self._cols = 0
                    self._rows = 0
                else: 
                    single = True
                    for i in args[0]:
                        if not isinstance(i, int) and not isinstance(i, float):
                            single = False
                    if single == True:
                        self._matrix = []
                        self._matrix.append(args[0])
                        self._cols = len(args[0])
                        self._rows = 1
                    else: 
                        if self._check(args[0]):
                            self._matrix = deepcopy(args[0])
                            self._rows = len(args[0])
                            self._cols = len(args[0][0])
                        else:
                            raise NameError('Undequate parameters for creation matrix with list parameter') 
            else:
                raise NameError('Undequate parameters for creation matrix (not list)') 
        elif len(args) > 1:
            if (not isinstance(args[0], int) or not isinstance(args[1], int)):
                raise NameError('Undequate parameters for creation matrix more than 1 parametes') 
            elif len(args) == 2:
                self._matrix = [[0 for x in range(args[1])] for y in range(args[0])]
                self._cols = args[1]
                self._rows = args[0]
            elif len(args) == 3:
                if isinstance(args[2], int) or isinstance(args[2], float):
                    self._matrix = [[args[2] for x in range(args[1])] for y in range(args[0])]
                    self._cols = args[1]
                    self._rows = args[0]
                else:
                    NameError('Undequate parameters for creation matrix more than 1 parametes')
            else:
                NameError('Undequate parameters for creation matrix more than 1 parametes')
    def __str__(self):
        res = []
        cur = ''
        for r in self._matrix:
            cur = ' '.join([str(item) for item in r])
            res.append(cur)
        return '\n'.join(res)

    def __repr__(self):
         return 'Matrix: ' + str(self._matrix)
    def __getitem__(self, i):
        return self._matrix[i]
    def __add__(self, other):
        if self._cols != other._cols or self._rows != other._rows:
            raise NameError("Can't sum this matrix")
        else:
            res = []
            cur = []
            for i in range(self._rows):
                for j in range(self._cols):
                    cur.append(self[i][j] + other[i][j])
                res.append(cur.copy())
                cur.clear()
            return Matrix(res)
    def transposition(self):
        return Matrix(list(map(list,[*zip(*self._matrix)])))
    def trace(self):
        if self._cols != self._rows:
            raise NameError("Can't calculate trace of this matrix")
        res = 0
        for row in range(self._rows):
            for col in range(self._cols):
                if row == col:
                    res += self._matrix[row][col]
        return res
    def _check(self, lst):
        size = len(lst[0])
        for i in lst:
            if not isinstance(i, list) or size != len(i):
                return False
            for e in i:
                if not isinstance(e, int) and not isinstance(e, float):
                    return False
        return True
    def __mul__(self, num):
        return Matrix(list(map(lambda x: list(map(lambda y: y * num, x)), self._matrix)))
    def __mul__(self, other):
        if (self._cols != other._rows):
            raise NameError("Can't multiply these matrices")
        res = Matrix(self._rows, other._cols)
        for rs in range(self._rows):
            for co in range(other._cols):
                for j in range(self._cols):
                    res[rs][co] += self[rs][j] * other[j][co]
        return res
    def __truediv__(self, num):
        return self.__mul__(1 / num)
    def __sub__(self, other):
        return self.__add__(other.__mul__(-1))
    def determinant(self):
        if self._cols != self._rows:
            raise NameError("Can't calculate determinant of this matrix")
        triangle_from = self.to_upper_triangle_form()
        det = 1
        for r in range(self._rows):
            for c in range(self._cols):
                if r == c:
                    det *= triangle_from[r][c]
        return det
    def to_upper_triangle_form(self):
        copy_of_matrix = deepcopy(self._matrix)
        for r in range(self._rows):
            for c in range(self._cols):
                if r == c:
                    if r != self._cols - 1:
                        for j in range(r, self._rows - 1):
                            if copy_of_matrix[r][c] != 0: 
                                tosub = list(map(lambda x: x * copy_of_matrix[j + 1][c] / copy_of_matrix[r][c], copy_of_matrix[r]))
                                new_row = []
                                for i in range(self._cols):
                                    new_row.append(copy_of_matrix[j + 1][i] - tosub[i])
                                copy_of_matrix[j + 1] = new_row.copy()
                                new_row.clear() 
        return Matrix(copy_of_matrix)
            
m = Matrix(5, 3, -1.5)
# print(m)
m1 = Matrix([[1.1, 2], [3, 4]])
# m2 = Matrix([[1], [2]])
# print(m1)
# print(m1 + m2)
# print('Initial matrix:')
# print(m1 * m2)
# print(m1.make_copy())
m1.write_to_file_text('Kek.txt')
m4 = Matrix.read_from_file_text('Kek.txt')
m1.write_to_file_bin('matrix.pkl')
m3 = Matrix.read_from_file_bin('matrix.pkl')
m4
# print('Transposed matrix:')
# print(m1.transposition())
# print(m2.trace())
# print(m1 - m2)
# print(m1.to_upper_triangle_form())
```

```python
class CoffeeInterface():
    def calculate_cost(self): raise NotImplementedError

class Latte(CoffeeInterface):
    def __init__(self):
        if hasattr(self, 'cost'):
            raise NameError("Impossible")
        setattr(self, 'cost', 10)
        
    def calculate_cost(self):
        return self.cost
    
class Cappuccino(CoffeeInterface):
    def __init__(self):
        if hasattr(self, 'cost'):
            raise NameError("Impossible")
        setattr(self, 'cost', 20)
        
    def calculate_cost(self):
        return self.cost
    
class Americano(CoffeeInterface):
    def __init__(self):
        if hasattr(self, 'cost'):
            raise NameError("Impossible")
        setattr(self, 'cost', 30)
        
    def calculate_cost(self):
        return self.cost
    
## additives in coffee

class Sugar(CoffeeInterface):
    def __init__(self, coffee):
        if not hasattr(coffee, 'cost'):
            raise NameError('Impossible')
        self.cost = coffee.cost + 1
    
    def calculate_cost(self):
        return self.cost
        
class Cream(CoffeeInterface):
    def __init__(self, coffee):
        if not hasattr(coffee, 'cost'):
            raise NameError('Impossible')
        self.cost = coffee.cost + 2
    def calculate_cost(self):
        return self.cost

class Cardamom (CoffeeInterface):
    def __init__(self, coffee):
        if not hasattr(coffee, 'cost'):
            raise NameError('Impossible')
        self.cost = coffee.cost + 3
    def calculate_cost(self):
        return self.cost

class Foam(CoffeeInterface):
    def __init__(self, coffee):
        if not hasattr(coffee, 'cost'):
            raise NameError('Impossible')
        self.cost = coffee.cost + 4
    def calculate_cost(self):
        return self.cost

class Syrup(CoffeeInterface):
    def __init__(self, coffee):
        if not hasattr(coffee, 'cost'):
            raise NameError('Impossible')
        self.cost = coffee.cost + 5
    def calculate_cost(self):
        return self.cost
## test
drink1 = Syrup(Foam(Sugar(Cardamom(Cream(Sugar(Latte()))))))
drink1.calculate_cost()
```
